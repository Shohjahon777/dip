```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

print("Libraries imported successfully!")
```

```python
img = np.zeros((20, 20), dtype=np.uint8)

img[5:16, 5:16] = 1

img[7:10, 7:10] = 0

img[12:15, 12:15] = 0

img[1, 10] = 1

img[19, 6] = 1

print(img)
```

```python
plt.figure(figsize=(5, 5))
plt.imshow(img, cmap='gray', vmin=0, vmax=1)
plt.title("Original Binary Image")
plt.axis("off")
plt.show()
```

```python
kernel = np.ones((5, 5), dtype=np.uint8)
print(kernel)
```

```python
eroded = cv2.erode(img, kernel, iterations=1)

plt.figure(figsize=(5, 5))
plt.imshow(eroded, cmap='gray', vmin=0, vmax=1)
plt.title("Eroded Image")
plt.axis("off")
plt.show()

print(eroded)
```

```python
dilated = cv2.dilate(img, kernel, iterations=1)

plt.figure(figsize=(5, 5))
plt.imshow(dilated, cmap='gray', vmin=0, vmax=1)
plt.title("Dilated Image")
plt.axis("off")
plt.show()
print(dilated)
```

```python
opened = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)

plt.figure(figsize=(5, 5))
plt.imshow(opened, cmap='gray', vmin=0, vmax=1)
plt.title("Opened Image")
plt.axis("off")
plt.show()
print(opened)
```

```python
closed = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(5, 5))
plt.imshow(closed, cmap='gray', vmin=0, vmax=1)
plt.title("Closed Image")
plt.axis("off")
plt.show()
print(closed)
```

```python
images = [img, eroded, dilated, opened, closed]
titles = ["Original", "Erosion", "Dilation", "Opening", "Closing"]

plt.figure(figsize=(15, 3))
for i in range(len(images)):
    plt.subplot(1, 5, i + 1)
    plt.imshow(images[i], cmap='gray', vmin=0, vmax=1)
    plt.title(titles[i])
    plt.axis("off")
plt.tight_layout()
plt.show()
```

```python
finger_img = cv2.imread('finger.png', cv2.IMREAD_GRAYSCALE)
_, binary_finger = cv2.threshold(finger_img, 0, 1, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.figure(figsize=(5, 5))
plt.imshow(binary_finger, cmap='gray', vmin=0, vmax=1)  
plt.title("Original Finger Image")
plt.axis("off")
plt.show()
```

```python
eroded_finger = cv2.erode(binary_finger, kernel, iterations=1)
plt.figure(figsize=(5, 5))
plt.imshow(eroded_finger, cmap='gray', vmin=0, vmax=1)
plt.title("Eroded Finger Image")
plt.axis("off")
plt.show()
```

```python
dilated_finger = cv2.dilate(binary_finger, kernel, iterations=1)
plt.figure(figsize=(5, 5))
plt.imshow(dilated_finger, cmap='gray', vmin=0, vmax=1)
plt.title("Dilated Finger Image")
plt.axis("off")
plt.show()
```

```python
opening_finger = cv2.morphologyEx(binary_finger, cv2.MORPH_OPEN, np.ones((4, 4), dtype=np.uint8))
plt.figure(figsize=(5, 5))
plt.imshow(opening_finger, cmap='gray', vmin=0, vmax=1)
plt.title("Opened Finger Image")
plt.axis("off")
plt.show()
```

```python
closing_finger = cv2.morphologyEx(binary_finger, cv2.MORPH_CLOSE, np.ones((4, 4)))
plt.figure(figsize=(5, 5)) 
plt.imshow(closing_finger, cmap='gray', vmin=0, vmax=1)
plt.title("Closed Finger Image")
plt.axis("off")
plt.show()
```

```python
def global_threshold(image, T):
    return (image >= T).astype(np.uint8)

T = 120
mask_global = global_threshold(finger_img, T)

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(finger_img, cmap="gray")
axes[0].set_title(f"Original image")
axes[0].axis("off")

axes[1].imshow(mask_global, cmap="gray")
axes[1].set_title(f"Global threshold mask (T={T})")
axes[1].axis("off")

plt.tight_layout()
plt.show()
```

```python
threshold_values = [60, 100, 140, 180, 220]

fig, axes = plt.subplots(1, len(threshold_values), figsize=(20, 4))

for ax, threshold in zip(axes, threshold_values):
    mask = global_threshold(closing_finger, threshold)
    foreground_pixels = int(mask.sum())
    foreground_percent = 100 * foreground_pixels / mask.size

    ax.imshow(mask, cmap="gray")
    ax.set_title(f"T={threshold}\n{foreground_percent:.1f}% foreground")
    ax.axis("off")

    print(
        f"Threshold {threshold}: foreground pixels = {foreground_pixels} "
        f"({foreground_percent:.1f}% of the image)"
    )

plt.tight_layout()
plt.show()

print("\nQuick observation:")
print("- Low thresholds keep more pixels as foreground, so the mask becomes larger and may include background.")
print("- High thresholds remove more pixels, so the mask becomes smaller and may miss parts of the object.")
print("- For this synthetic image, T=140 is usually a good balance between object and background separation.")
```

```python
text_img = cv2.imread('text.png', cv2.IMREAD_GRAYSCALE)

_, binary_text = cv2.threshold(text_img, 0, 1, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
plt.figure(figsize=(5, 5))
plt.imshow(binary_text, cmap='gray', vmin=0, vmax=1)
plt.title("Otsu's Thresholding Result")
plt.axis("off")
plt.show()
```

```python
opening_text = cv2.morphologyEx(binary_text, cv2.MORPH_OPEN, np.ones((1, 1), dtype=np.uint8))
plt.figure(figsize=(5, 5))
plt.imshow(opening_text, cmap='gray', vmin=0, vmax=1)
plt.title("Opened Text Image")
```

```python
dilation_text = cv2.dilate(binary_text, np.ones((3, 3), dtype=np.uint8), iterations=1)
plt.figure(figsize=(5, 5))
plt.imshow(dilation_text, cmap='gray', vmin=0, vmax=1)
plt.title("Dilated Text Image")
plt.axis("off")
plt.show()
```

```python
closed_text = cv2.morphologyEx(binary_text, cv2.MORPH_CLOSE, np.ones((3, 3), dtype=np.uint8))
plt.figure(figsize=(5, 5))
plt.imshow(closed_text, cmap='gray', vmin=0, vmax=1)
plt.title("Closed Text Image")
plt.axis("off")
plt.show()
```
