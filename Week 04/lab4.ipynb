```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


def show(title, img, cmap=None):
    plt.figure(figsize=(5,4))
    if cmap:
        plt.imshow(img, cmap=cmap)
    elif len(img.shape) == 3:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(img, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
def show_two(title1, img1, title2, img2):
    fig, axes = plt.subplots(1, 2, figsize=(10, 4))
    for ax, img, title in zip(axes, [img1, img2], [title1, title2]):
        if len(img.shape) == 3:
            ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        else:
            ax.imshow(img, cmap='gray')
        ax.set_title(title, fontsize=13)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
    
print('Libraries loaded successfully')
```

```python
def color_to_gray(image_path):
    img = cv2.imread(image_path)
    
    if img is None:
        print(f"Could not load image: {image_path}")
        return None
    
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    show_two('Input (COLOR)', img, 'OUTPUT (GRAYSCALE)', gray)
    return gray

color_to_gray("D:\DIP\DIP_WORKS\images.jpg")
```

```python
def mirror_effect(image_path):
    
    img = cv2.imread(image_path)
    
    if img is None:
        print("No image provided")
        return None
    flipped = cv2.flip(img, 1)
    mirrored = np.hstack([img, flipped])
    
    show_two('Input Image', img, "Mirrored Output", mirrored)
    show_two('Input Image', img, "Flipped Output", flipped)
    return mirrored


mirror_effect("D:\DIP\DIP_WORKS\images.jpg")
```

```python
def image_negative(image_path):

    img = cv2.imread(image_path)
    
    if img is None:
        print(f" Could not load: {image_path}")
        return None
    
    negative = 255 - img
    
    
    show_two('Original', img, 'Negative (255 - r)', negative)
    return negative


neg_result = image_negative('D:\DIP\DIP_WORKS\images.jpg')

print('\n Pixel value examples:')
print(f'  Black (0)    {255 - 0}   (White)')
print(f'  White (255)  {255 - 255}   (Black)')
print(f'  Gray  (128)  {255 - 128}  (Light gray)')
print(f'  Dark  (50)   {255 - 50}  (Bright)')
```

```python
# ============================================================
# Bit-Plane Slicing: displayBitResponse
# ============================================================

def displayBitResponse(image_path):
    """
    Extracts and displays all 8 bit planes of a grayscale image.
    
    How to extract bit n:
      - Right-shift pixel value by (n-1) positions
      - AND with 1 to keep only that bit
      - Formula: bit_plane = (img >> (n-1)) & 1
    """
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    
    if img is None:
        print(f" Could not load: {image_path}")
        return
    
    fig, axes = plt.subplots(2, 4, figsize=(16, 8))
    axes = axes.flatten()

    for bit in range(1, 9):   # Bits 1 through 8
        # Extract bit plane:
        # >> (bit-1) shifts right by (bit-1) positions
        # & 1 keeps only the lowest bit (0 or 1)
        bit_plane = (img >> (bit - 1)) & 1
        
        # Scale 0/1  0/255 for visibility
        bit_plane_display = bit_plane * 255

        axes[bit-1].imshow(bit_plane_display, cmap='gray', vmin=0, vmax=255)
        axes[bit-1].set_title(f'Bit Plane {bit}  (weight={2**(bit-1)})', fontsize=11)
        axes[bit-1].axis('off')

    plt.suptitle('All 8 Bit Planes\n(Plane 1=LSB/least info  Plane 8=MSB/most info)', 
                 fontsize=14)
    plt.tight_layout()
    plt.show()


# ---- TEST IT ----
# Create a more interesting test image with varied intensities
test_bp = np.zeros((200, 200), dtype=np.uint8)
test_bp[0:50, :]   = 200   # Top row: 200
test_bp[50:100, :] = 128   # Second: 128 (10000000 in binary)
test_bp[100:150, :] = 64   # Third: 64
test_bp[150:200, :] = 255  # Bottom: 255
# Add some pattern
test_bp[::10, ::10] = 0
cv2.imwrite('bitplane_test.jpg', test_bp)

displayBitResponse('D:\DIP\DIP_WORKS\images.jpg')
```
