```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image (RGB/BGR)
image = np.zeros((512, 512, 3), np.uint8)
# (512, 512, 3): height × width × 3 color channels (BGR)
# np.uint8: 8-bit unsigned integer type (pixel values 0–255)

# Create a black & white (grayscale) image
image_bw = np.zeros((512, 512), np.uint8)
# Only 2D because grayscale has a single channel

# Display color image (convert BGR → RGB for matplotlib)
plt.imshow(image[..., ::-1])
plt.title("Black Color Image")

# Display grayscale image
plt.figure()
plt.imshow(image_bw, cmap='gray')
plt.title("Black & White Image")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image
image = np.zeros((512, 512, 3), np.uint8)

# Draw a diagonal line from top-left to bottom-right
cv2.line(image, (0, 0), (511, 511), (255, 127, 0), 5)
# (255, 127, 0) is the color in BGR (Blue, Green, Red)
# 5 is the thickness of the line in pixels

# Display using OpenCV window
cv2.imshow("Blue Line", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display using matplotlib
plt.imshow(image[..., ::-1])  # Convert BGR to RGB for matplotlib
plt.title("Diagonal Line")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image
image = np.zeros((512, 512, 3), np.uint8)

# Draw a rectangle
cv2.rectangle(image, (100, 100), (300, 250), (0, 0, 255), 5)
# (100, 100) → top-left corner
# (300, 250) → bottom-right corner
# (0, 0, 255) → red color in BGR
# 5 → thickness of the rectangle border

# Display using OpenCV
cv2.imshow("Rectangle", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display using Matplotlib
plt.imshow(image[..., ::-1])  # Convert BGR → RGB
plt.title("Rectangle")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image
image = np.zeros((512, 512, 3), np.uint8)

# Draw a circle
cv2.circle(image, (350, 350), 100, (15, 75, 50), 7)
# (350, 350) → center of the circle
# 100 → radius in pixels
# (15, 75, 50) → color in BGR (Blue, Green, Red)
# 7 → thickness of the circle border

# Display using OpenCV
cv2.imshow("Circle", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display using Matplotlib
plt.imshow(image[..., ::-1])  # Convert BGR → RGB
plt.title("Circle")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image
ellipse = np.zeros((512, 512, 3), np.uint8)

# Draw a filled ellipse
cv2.ellipse(ellipse, (250, 250), (250, 100), 45, 90, 360, (255, 255, 255), -1)
# (250, 250) → center of the ellipse
# (250, 100) → lengths of the major and minor axes
# 45 → rotation angle in degrees
# 90 → starting angle of the arc
# 360 → ending angle of the arc
# (255, 255, 255) → color in BGR (white)
# -1 → fill the ellipse completely

# Display using OpenCV
cv2.imshow("Ellipse", ellipse)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display using Matplotlib
plt.imshow(ellipse[..., ::-1])  # Convert BGR → RGB
plt.title("Ellipse")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image
image = np.zeros((512, 512, 3), np.uint8)

# Define four points of the polygon
pts = np.array([[10, 50], [400, 50], [90, 200], [50, 500]], np.int32)

# Reshape points for cv2.polylines: required shape (-1, 1, 2)
pts = pts.reshape((-1, 1, 2))

# Draw the polygon
cv2.polylines(image, [pts], True, (0, 0, 255), 3)
# True → close the polygon
# (0, 0, 255) → red color in BGR
# 3 → thickness of the lines

# Display using OpenCV
cv2.imshow("Polygon", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display using Matplotlib
plt.imshow(image[..., ::-1])  # Convert BGR → RGB
plt.title("Polygon")
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Create a black color image
image = np.zeros((512, 512, 3), np.uint8)

# Add text to the image
cv2.putText(image, 'Hello World!', (75, 290), cv2.FONT_HERSHEY_COMPLEX, 2, (100, 170, 0), 3)
# 'Hello World!' → text to display
# (75, 290) → bottom-left corner of the text
# cv2.FONT_HERSHEY_COMPLEX → font type
# 2 → font scale (size)
# (100, 170, 0) → text color in BGR
# 3 → thickness of the text strokes

# Display using OpenCV
cv2.imshow("Hello World!", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Display using Matplotlib
plt.imshow(image[..., ::-1])  # Convert BGR → RGB
plt.title("Hello World!")
plt.show()
```
